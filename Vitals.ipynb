```python
from datetime import datetime
from dateutil.parser import parse
import pandas as pd
```

```python
df= pd.read_csv('/Users/jmarquez/Dropbox/Current Documents/Programming/Vitals.csv', header=0)
df.head()
```

```python
df.info()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], errors='coerce')
df['TIME'] = pd.to_datetime(df['TIME'], errors='coerce', \
                            format = '%H:M').dt.time
df.info()
```

```python
df.head()
```

```python
hi_hr =[]

for row in df['HR']:
    if row >100:
        hi_hr.append(1)
    else:
        hi_hr.append(0)
df['hi_hr']= hi_hr
```

```python
low_bp=[]

for row in df['SBP']:
    if row <100:
        low_bp.append(1)
    else:
        low_bp.append(0)
        
df['low_bp'] = low_bp

df.head()
```

```python
hi_RR=[]

for row in df['RR']:
    if row >20:
        hi_RR.append(1)
    else:
        hi_RR.append(0)

df['hi_rr']= hi_RR

df.head()
```

```python
hi_wbc= []

for row in df['WBC']:
    if row > 11000:
        hi_wbc.append(1)
    else:
        hi_wbc.append(0)

df['hi_WBC']= hi_wbc

df.head()
```

```python
hi_temp= []

for row in df['TEMP']:
    if row > 100:
        hi_temp.append(1)
    else:
        hi_temp.append(0)
    
df['hi_temp']= hi_temp

df.head()
```

```python
df['sirs_score']=df.loc[:,'hi_hr':'hi_temp'].sum(axis=1)
df
```
